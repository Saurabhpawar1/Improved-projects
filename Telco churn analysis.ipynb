## Data cleaning

1. Create a copy of base data for manupulation & processing

```python
telco_data = telco_base_data.copy()
```

2. Total Charges should be numeric amount. Let's convert it to numerical data type

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

3. As we can see there are 11 missing values in TotalCharges column. Let's check these records

```python
telco_data.loc[telco_data ['TotalCharges'].isnull() == True]
```

## Missing value treatment

Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.

```python
#Removing missing values 
telco_data.dropna(how = 'any', inplace = True)
```

5. Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
# Get the max tenure
print(telco_data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

6. Remove columns not required for processing

```python
#drop column customerID and tenure
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telco_data.head()
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

# Model building

```python
x=telco_data_dummies.drop('Churn',axis=1)
x
```

```python
y=telco_data_dummies['Churn']
y
```

```python
#splitting the data into training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

## Handling imbalance data through SMOTEENN technique

```python
from imblearn.combine import SMOTEENN 
sm = SMOTEENN()
x_resampled,y_resampled=sm.fit_sample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test=train_test_split(x_resampled, y_resampled,test_size=0.2)
```

```python
#Importing all the model library

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

#Importing Boosting models
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier

#Importing error metrics
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
KNN=KNeighborsClassifier(n_neighbors=6)
LR=LogisticRegression()
DT=DecisionTreeClassifier(random_state=6)
XGB=XGBClassifier()
RF=RandomForestClassifier()
ADA=AdaBoostClassifier()
GNB=GaussianNB()
GBC=GradientBoostingClassifier()
BC=BaggingClassifier()
ETC=ExtraTreesClassifier()
```

```python
models= []
models.append(('KNeighborsClassifier', KNN))
models.append(('LogisticRegression', LR))
models.append(('DecisionTreeClassifier', DT))
models.append(('XGBClassifier', XGB))
models.append(('RandomForestClassifier', RF))
models.append(('AdaBoostClassifier', ADA))
models.append(('GaussianNB', GNB))
models.append(('GradientBoostingClassifier', GBC))
models.append(('BaggingClassifier', BC))
models.append(('ExtraTreesClassifier', ETC))
```

```python
Model= []
score= []
cvs=[]
rocscore=[]
for name,model in models:
    print('******************************************',name,'********************************************************')
    print('\n')
    Model.append(name)
    model.fit(xr_train, yr_train)
    print(model)
    pre=model.predict(xr_test)
    print('\n')
    AS=accuracy_score(yr_test,pre)
    print('Accuracy_score = ',AS)
    score.append(AS*100)
    print('\n')
    sc= cross_val_score(model, x_resampled,y_resampled, cv=10, scoring='accuracy').mean()
    print('Cross_Val_Score = ',sc)
    cvs.append(sc*100)
    print('\n')
    false_positive_rate, true_positive_rate, thresholds = roc_curve(yr_test,pre)
    roc_auc= auc(false_positive_rate, true_positive_rate)
    print ('roc_auc_score = ',roc_auc)
    rocscore.append(roc_auc*100)
    print('\n')
    print('classification_report\n',classification_report(yr_test,pre))
    print('\n')
    cm=confusion_matrix(yr_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate, true_positive_rate, label='AUC = %0.2f'% roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    print('\n\n')
```

```python
result = pd.DataFrame({'Model': Model, 'Accuracy_score': score,'Cross_val_score': cvs,'Roc_auc_curve':rocscore})
result
```

## XGBClassifier is giving us best rocscore

# Hyperparameter tuning

```python
parameters={'n_estimators': [100, 250, 500], 'max_depth': [6,9,12], 'subsample':[0.9, 1.0]}
XGB=XGBClassifier()

clf=GridSearchCV(XGB, parameters, n_jobs=-1)
clf.fit(xr_train,yr_train)
print(clf.best_params_)
```

```python
#XGBClassifier with best parameters

XGB=XGBClassifier(max_depth=6,n_estimators=100,subsample=0.9,random_state=9)
XGB.fit(xr_train,yr_train)
XGB.score(xr_test,yr_test)
predXGB=XGB.predict(xr_test)
print(accuracy_score(yr_test,predXGB))
print(confusion_matrix(yr_test,predXGB))
print(classification_report(yr_test,predXGB))
```

```python
#cross validate XGBClassifier accuracy

score=cross_val_score(XGB,x_resampled,y_resampled,cv=5,scoring='accuracy')

print(score)
print('Mean:',score.mean())
print('Std:',score.std())
```

```python
#plotting the roc auc curve

from sklearn import metrics
y_pred_proba = XGB.predict_proba(xr_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(yr_test,  y_pred_proba)
auc = metrics.roc_auc_score(yr_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
